```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
from pathlib import Path
sns.set_style("whitegrid")
```

```python
demographics=pd.read_csv('./demographics/demographics_cleaned.csv')
demographics.info()
```

```python
#better_demographics=demographics[['GEO_ID','NAME']]
```

```python
demographics['pop_18_30']=demographics['pop_18_24']+demographics['pop_25_30']
demographics['pop_18_30_pc']=100*demographics['pop_18_30']/demographics['total_pop']
demographics['pop_60_up']=demographics['pop_60_65']+demographics['pop_65_70']+demographics['pop_70_75']+demographics['pop_75_80']+demographics['pop_80_85']+demographics['pop_85_90']
demographics['pop_60_up_pc']=100*demographics['pop_60_up']/demographics['total_pop']

demographics['pop_male_pc']=100*demographics['total_male_pop']/demographics['total_pop']
final_tally=demographics[['GEO_ID','NAME','total_pop','pop_18_30_pc','pop_60_up_pc','pop_male_pc']]
```

```python
final_tally.set_index('GEO_ID',inplace=True)
```

```python
final_tally.info()
```

```python
dp_2016=pd.read_csv('./demographics/dp_2016_cleaned.csv')
dp_2019=pd.read_csv('./demographics/dp_2019_cleaned.csv')
dp_2016.drop(axis=1,labels=['Unnamed: 0'],inplace=True)
dp_2019.drop(axis=1,labels=['Unnamed: 0'],inplace=True)
dp_2016.set_index('GEO_ID',inplace=True)
dp_2019.set_index('GEO_ID',inplace=True)
```

```python
dp_2019.loc[:,'tot_pop']
```

```python
delta_dp=dp_2016[['NAME']]
for x in dp_2016.columns:
    if x != 'NAME':
        dp_2016[x+'_pc']=100*dp_2016[x]/dp_2016['tot_pop']
        dp_2019[x+'_pc']=100*dp_2019[x]/dp_2019['tot_pop']
        delta_dp.loc[:,x+'_delta']=dp_2019[(x+'_pc')]-dp_2016[(x+'_pc')]
delta_dp.drop(axis=1,labels=['tot_pop_delta'],inplace=True)
```

```python
cols=['afr_amer','amer_ind','asian','latino','white']
final_tally.loc[:,[x+'_pc' for x in cols]]=dp_2019.loc[:,[ x+'_pc' for x in cols]]
final_tally.loc[:,[x+'_delta' for x in cols]]=delta_dp.loc[:,[x+'_delta' for x in cols]]
```

```python
final_tally.info()
```

```python
poverty_2019=pd.read_csv('./poverty/poverty_2019_cleaned.csv',engine='python')
poverty_2016=pd.read_csv('./poverty/poverty_2016_cleaned.csv',engine='python')
poverty_2016.set_index('GEO_ID',inplace=True)
poverty_2019.set_index('GEO_ID',inplace=True)
poverty_2016.drop(axis=1,labels=['Unnamed: 0'],inplace=True)
poverty_2019.drop(axis=1,labels=['Unnamed: 0'],inplace=True)

```

```python
poverty_2016['pov_pc']=100*poverty_2016['tot_poverty']/final_tally['total_pop']
poverty_2019['pov_pc']=100*poverty_2019['tot_poverty']/final_tally['total_pop']
poverty_2016['pov_150_pc']=100*poverty_2016['pov_150']/final_tally['total_pop']
poverty_2019['pov_150_pc']=100*poverty_2019['pov_150']/final_tally['total_pop']
```

```python
final_tally['pov_pc']=poverty_2019['pov_pc']
final_tally['pov_pc_delta']=poverty_2019['pov_pc']-poverty_2016['pov_pc']
```

```python
final_tally[:5]
```

```python
income_2019=pd.read_csv('./income/income_2019_cleaned.csv',engine='python')
income_2016=pd.read_csv('./income/income_2016_cleaned.csv',engine='python')
income_2016.set_index('GEO_ID',inplace=True)
income_2019.set_index('GEO_ID',inplace=True)
income_2016.drop(axis=1,labels=['Unnamed: 0'],inplace=True)
income_2019.drop(axis=1,labels=['Unnamed: 0'],inplace=True)
```

```python
urban_rural=pd.read_csv('./urban_rural/urban_rural.csv')
urban_rural=urban_rural.loc[1:]
urban_rural.set_index('GEO_ID',inplace=True)
urban_rural['P2_001N']=urban_rural['P2_001N'].apply(int)
urban_rural['P2_002N']=urban_rural['P2_002N'].apply(int)
urban_rural['P2_003N']=urban_rural['P2_003N'].apply(int)

#urban_rural['urban']= 1 if urban_rural['P2_002N']
final_tally['urban']=100*urban_rural['P2_002N']/urban_rural['P2_001N']
final_tally['rural']=100*urban_rural['P2_003N']/urban_rural['P2_001N']
```

```python
final_tally['unemp_rate']=income_2019['unemp_rate']
final_tally['unemp_delta']=income_2019['unemp_rate']-income_2016['unemp_rate']
final_tally['mean_hhi']=income_2019['mean_hh_inc']
final_tally['med_hhi']=income_2019['med_hh_inc']
```

```python
education=pd.read_csv('./education/education_2019_cleaned.csv')
education.set_index('GEO_ID',inplace=True)
education.drop(axis=1,labels=['Unnamed: 0'],inplace=True)
```

```python
final_tally['hs_deg']=education['high_school_deg']
final_tally['bac_deg']=education['bachelor_deg']
```

```python
r='REPUBLICAN'
d='DEMOCRAT'
election1=pd.read_csv('./election_results/countypres_2020.csv')
election=election1.loc[(election1['year']==2020) & ((election1['party']==r)|(election1['party']==d))]
election_old=election1.loc[(election1['year']==2016) & ((election1['party']==r)|(election1['party']==d))]
election.dropna(inplace=True)
election_old.dropna(inplace=True)
```

```python
election_old[:5]
```

```python
election['GEO_ID']=election['county_fips'].apply(lambda x:'0500000US'+str(int(x)).zfill(5))
election_old['GEO_ID']=election_old['county_fips'].apply(lambda x:'0500000US'+str(int(x)).zfill(5))
```

```python
elections_pc=100*election.groupby(['GEO_ID','party'])['candidatevotes'].sum()/election.groupby(['GEO_ID','party'])['totalvotes'].mean()
elections_old_pc=100*election_old.groupby(['GEO_ID','party'])['candidatevotes'].sum()/election_old.groupby(['GEO_ID','party'])['totalvotes'].mean()
```

```python
elections_old_pc[:5]
```

```python
for x in final_tally.index.intersection(election['GEO_ID']):
    final_tally.loc[x,'democrat']=elections_pc[x][d]
    final_tally.loc[x,'republican']=elections_pc[x][r]
for x in final_tally.index.intersection(election_old['GEO_ID']):
    final_tally.loc[x,'democrat_old']=elections_old_pc[x][d]
    final_tally.loc[x,'republican_old']=elections_old_pc[x][r]
```

```python
final_tally=final_tally.dropna()
```

```python
final_tally.to_csv('../aggregate.csv')
```
